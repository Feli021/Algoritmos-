# Desafios do HackerRank - Semana 1

Esta pasta contém os desafios resolvidos da Semana 1 do HackerRank. Cada desafio foca em habilidades específicas, com o objetivo de melhorar a compreensão de algoritmos e estruturas de dados.

## Desafios e Objetivos

### 1. **Plus_Minus**
**Objetivo**: Calcular e exibir a proporção de números positivos, negativos e zeros em um array, com precisão de 6 casas decimais.

**Habilidade Principal**: Manipulação de arrays e cálculos de proporções.

---

### 2. **Lonely_Integer**
**Objetivo**: Encontrar e retornar o único número que aparece apenas uma vez em um array, enquanto os outros aparecem em pares.

**Habilidade Principal**: Manipulação de arrays e identificação de elementos únicos.

---

### 3. **Mini_Max_Sum**
**Objetivo**: Calcular a menor e maior soma possíveis, excluindo um único valor do array, e exibir os resultados.

**Habilidade Principal**: Manipulação de arrays e cálculos de soma.

---

### 4. **Time_Conversion**
**Objetivo**: Converter uma hora no formato de 12 horas para o formato de 24 horas.

**Habilidade Principal**: Manipulação de strings e conversão de formatos de tempo.

---

### 5. **Breaking_Records**
**Objetivo**: Contar o número de vezes que os recordes de maior e menor pontuação foram quebrados em uma sequência de pontuações.

**Habilidade Principal**: Manipulação de arrays e contagem de recordes.

---

### 6. **Divisible_Sum_Pairs**
**Objetivo**: Contar o número de pares de elementos em um array cuja soma seja divisível por um número \( k \).

**Habilidade Principal**: Manipulação de arrays e operações matemáticas.

---

### 7. **Camel_Case_Split_And_Combine**
**Objetivo**: Manipular strings em CamelCase, dividindo-as em palavras e, em seguida, combinando palavras separadas de volta para o formato CamelCase, com a consideração de tipos como variável, método ou classe.

**Habilidade Principal**: Manipulação de strings e transformação de formatos.

---
